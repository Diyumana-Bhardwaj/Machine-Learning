```python
import pandas as pd
import random
```

```python
outcomes=[]
for i in range (1000):
    outcomes.append(random.randint(1,6))
```

```python
outcomes[:10]
```

```python
out=pd.Series(outcomes)
print(out.value_counts())
print(out.value_counts().sum())
```

```python
PX=(out.value_counts()/out.value_counts().sum()).sort_index()
PX
```

```python
print((out.value_counts()/out.value_counts().sum()).cumsum())
print((out.value_counts()/out.value_counts().sum()).sum())
```

```python
PX.plot(kind='bar')
```

```python
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt

sample=normal(loc=50,scale=10,size=1000)
sample
```

```python
plt.hist(sample)
```

firstly we generate some random samples then we plot a histogram of those samples upon seeing the histogram we try to analyse or try to match with some kind of distributions we know. ones we found or assumed a particular distribution we will take its parameters calculate then a put into its PDF function.



---


NOW AS WE CAN SEE THE ABOVE DISTRIBUTION IS SIMILAR TO NORMAL DISTRIBUTION WE
WILL CALCULATE ITS PARAMETERS SYCH AS MEAN AND STANDARD VARIENCE

```python
sample_mean= sample.mean()
sample_std=sample.std()
print(sample_mean)
print(sample_std)
```

now on next step will be fitting the distribution with the above calculated parameters matlab jo Humne parameters calculate Kiye Hain we already mentioned them while generating the data we already know those parameters but we are just assuming right now that we calculated the parameters upon seen the histogram and the type of distribution created above now fitting those parameters into our
samples.

```python
from scipy.stats import norm
dist=norm(sample_mean,sample_std)
dist
```

this dist is nothing but the function we created for the particular value of mean and standard deviation in the PDF function for normal distribution



---

Ab Jo sample Humne generate kiya tha usmein se Ham 100 equally space values nikaalenge Uske minimum aur maximum values ki range me aur unn values ko Hamne Jo PDF function create Ki Hai Upar with the particular value of mean and standard deviation usmein put kar denge

```python
#generating 100 equally spaced values from our sample
values=np.linspace(sample.min(),sample.max(),100)
```

```python
pdf=[dist.pdf(value) for value in values]
```

```python
plt.hist(sample,bins=10,density=True) #density =True becoz both have diffrent scale
plt.plot(values,pdf)
```

```python
sample1=normal(loc=20,scale=5,size=300)
sample2=normal(loc=40,scale=5,size=700)
# sample=sample1+sample2 will generate error
sample=np.concatenate([sample1,sample2])
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity
```

```python
#creating object of KernelDensity
node1=KernelDensity(kernel='gaussian',bandwidth=2) # change bandwidth to change the shape of grapgh
#lower the bandwidth more precise the graph is
#kernel will be distribution type and bandwidth will be standard deviation which will tell skewness

#converting data to 2D
sample=sample.reshape(len(sample),1)

node1.fit(sample) # error coz mostly sklearn lib work on 2D data
```

```python
values=np.linspace(sample.min(),sample.max(),100)
values=values.reshape(len(values),1)
```

```python
probablilities=node1.score_samples(values)
probablilities=np.exp(probablilities)
```

score_sample(values) returns the log-density estimate of the input sample values. This is because score_sample() method returns the logarithm of the probability density estimation rather than actual density estimation. Therefore to calculation actual density estimation exponential function is used.

```python
plt.hist(sample,bins=50,density=True)
plt.plot(values,probablilities)
plt.show()
```

```python
# prompt: cdf of sample without inbuilt function

import numpy as np
import matplotlib.pyplot as plt

def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n + 1) / n

    return x, y

# Example usage (replace with your actual sample data)
sample = np.random.normal(loc=20, scale=5, size=300)  # Example sample data

x, y = ecdf(sample)

# Plot the ECDF
plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('Data Values')
plt.ylabel('ECDF')
plt.title('Empirical Cumulative Distribution Function (ECDF)')
plt.grid(True)
plt.show()
```
